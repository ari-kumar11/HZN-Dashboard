```python
import gspread
import pandas as pd
import matplotlib.pyplot as plt

# Use your service account credentials
gc = gspread.service_account(filename='C:/Users/arind/Downloads/hzn-dashboard-406918-9134aab497bf.json')

# Open the Google Spreadsheet using the URL
sh = gc.open_by_url('https://docs.google.com/spreadsheets/d/1Z5Z8YEk4xk58IzQPFL83xPRKQWDN6WOv5YdnQakf4EE')

# Select the first sheet
worksheet = sh.get_worksheet(0)

# Get all data from the sheet
data = worksheet.get_all_values()

# Convert the data to a DataFrame
df = pd.DataFrame(data[1:], columns=data[0])

# Define the categories to create pie charts for
categories = ['Opensource', 'Stage', 'Gender', 'Technical Background', 
              'Time Commitment (Hours per week)', 'How did they hear about HZN', 'Round 1 Consensus']

# Create a figure to hold all pie charts
fig, axs = plt.subplots(len(categories), figsize=(10, 6 * len(categories)))

for i, category in enumerate(categories):
    # Count the occurrences of each category
    counts = df[category].value_counts()

    # Create a pie chart
    axs[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Distribution of {category}')

plt.tight_layout()
plt.show()
```
